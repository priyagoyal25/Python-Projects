# Text to image Generator using openai API

```python
import openai
import requests
from PIL import Image

openai.api_key = 'Your API here'  # Replace your openai API here

def generate(text):
    try:
        res = openai.Image.create(prompt=text, n=1, size="512x512")
        return res["data"][0]["url"]
    except Exception as e:
        print("Error generating image:", e)
        return None

text = "a lion with family"
url1 = generate(text)

if url1:
    try:
        response = requests.get(url1)
        if response.status_code == 200:
            with open("openai_image.jpg","wb") as file:
                file.write(response.content)
            img = Image.open("openai_image.jpg")
            img.show()
        else:
            print("Error retrieving image:", response.status_code)
    except Exception as e:
        print("Error opening image:", e)
        
```
