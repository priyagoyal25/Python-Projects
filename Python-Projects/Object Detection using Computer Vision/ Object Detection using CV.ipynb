# Object Detection using Computer Vision

This Python script utilizes computer vision techniques to detect and count objects of a specified type within an image. It employs the `cvlib` library, which is built on top of OpenCV for object detection. The script prompts the user to input an image path and the name of the object to be detected, and then it displays the image with bounding boxes drawn around the detected objects of that type.

4. **View the Results:**

## Dependencies

This script relies on the following Python libraries:

- `cv2`: OpenCV for image processing.
- `numpy`: For numerical operations on arrays.
- `matplotlib.pyplot`: Used for displaying the image.
- `cvlib`: A high-level computer vision library built on top of OpenCV for object detection.

## Object Detection

The script uses the `cvlib` library to perform object detection on the input image. It first reads the image, detects common objects within it, and counts the number of objects of the specified type.

## Example Usage

You can use this script to quickly identify and count objects of interest within images. For example, you could use it to count the number of cars in a traffic scene, detect animals in wildlife images, or find specific objects in photographs.

Feel free to modify and extend the script to suit your specific use case or integrate it into a larger computer vision application.

````python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import cvlib as cv
from cvlib.object_detection import draw_bbox
from numpy.lib.polynomial import poly
imgpath=input("Enter the path of image: ")
obj=input("Enter the name of object you want to find: ")
img=cv2.imread(imgpath)
box, label, count = cv.detect_common_objects(img)
output=draw_bbox(img,box,label,count)
plt.figure(figsize=[12,12])
plt.imshow(output[:,:,::-1])
plt.show()
print(f"Number of {obj}s in this image are: {str(label.count(obj))}")
````
