```python
import tkinter as tk
from tkinter import *

def encrypt():
    message = message_entry.get(1.0,END)
    key = key_entry.get()
    
    if key != "p2506":
        messagebox.showerror("Error", "Wrong Password")
        return

    encrypted_message = ""
    key_length = len(key)
    for i, char in enumerate(message):
        key_char = key[i % key_length]
        shift = ord(key_char) - ord('A')  # Shift value based on the key character
        
        if char.isalpha():
            if char.islower():
                encrypted_char = chr(((ord(char) - ord('a') + shift) % 26) + ord('a'))
            else:
                encrypted_char = chr(((ord(char) - ord('A') + shift) % 26) + ord('A'))
        else:
            encrypted_char = char
        
        encrypted_message += encrypted_char
    
    result_label.config(text=f"Encrypted message: {encrypted_message}")

def decrypt():
    message = message_entry.get(1.0,END)
    key = key_entry.get()
    
    if key!= "p2506":
        messagebox.showerror("Error", "Wrong Password ")
        return

    decrypted_message = ""
    key_length = len(key)
    for i, char in enumerate(message):
        key_char = key[i % key_length]
        shift = ord(key_char) - ord('A')  # Shift value based on the key character
        
        if char.isalpha():
            if char.islower():
                decrypted_char = chr(((ord(char) - ord('a') - shift + 26) % 26) + ord('a'))
            else:
                decrypted_char = chr(((ord(char) - ord('A') - shift + 26) % 26) + ord('A'))
        else:
            decrypted_char = char
        
        decrypted_message += decrypted_char
    
    result_label.config(text=f"Decrypted message: {decrypted_message}")

# Create the main window
root = tk.Tk()
root.geometry("600x450")
root.title("Encrypt Decrypt")

# Create labels and entry widgets with font size 16
message_label = tk.Label(root, text="Enter the message:", font=("Arial", 16))
message_entry = tk.Text(root, width=45,height=5, font=("Arial", 14))
key_label = tk.Label(root,height=2, text="Enter the alphanumeric key:", font=("Arial", 16))
key_entry = tk.Entry(root,width=45, font=("Arial", 14),show=".")

# Create buttons with different colors and font size 16
encrypt_button = tk.Button(root, text="Encrypt", command=encrypt, height=1, width=15, bg="orange", fg="white", bd=0, font=("Arial", 16))
decrypt_button = tk.Button(root, text="Decrypt",  command=decrypt, height=1, width=15, bg="green", fg="white", bd=0, font=("Arial", 16))

# Create result label with font size 16
result_label = tk.Label(root, text="", fg="blue", font=("Arial", 16))

# Arrange widgets using grid layout
message_label.grid(row=0, column=0)
message_entry.grid(row=1, column=0, columnspan=2)
key_label.grid(row=2, column=0)
key_entry.grid(row=3, column=0, columnspan=2)
encrypt_button.grid(row=4, column=0,padx=5, pady=10,)
decrypt_button.grid(row=4, column=1,padx=5, pady=10,)
result_label.grid(row=6, column=0, columnspan=2)

# Function to reset the fields
# def reset():
#     message_entry.delete(0, tk.END)
#     key_entry.delete(0, tk.END)
#     result_label.config(text="")
def reset():
    message_entry.delete(1.0,END)
    key_entry.delete(0, END)
    result_label.config(text="")
# Create a Reset button
reset_button = tk.Button(root, text="RESET", height=2, width=50, bg="#1089ff", fg="white", bd=0, command=reset, font=("Arial", 16))
reset_button.grid(row=5, column=0, padx=5, pady=10, columnspan=2)

root.mainloop()
```
